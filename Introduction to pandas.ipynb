### basic function and filter in pandas

#### Basic Functions in Pandas

`mean()`: Compute mean of groups

`sum()`: Compute sum of group values

`size()`: Compute group sizes

`count()`: Compute count of group

`std()`: Standard deviation of groups

`var()`: Compute variance of groups

`min()`: Compute min of group values

`unique()`: Return unique values of group

`nunique()`: Return number of unique values of group

`value_counts()`: Count frequency of unique values of group

```python
df['Survived']
```

```python
df['Survived'].mean()
```

```python
df['Survived'].count()
```

```python
df['Survived'].value_counts()
```

#### filter

```python
df[df['Survived'] == 1]
```

```python
df[df['Age'] > 20]
```

```python
df[(df['Age'] > 20) & (df['Survived'] == 1)]
```

#### group by and pivot table

```python
df.groupby('Sex').mean()
```

```python
df.groupby(['Sex', 'Pclass']).mean()
```

```python
df.groupby(['Sex', 'Pclass'])['Survived'].agg(['mean', 'sum'])
```

```python
pd.pivot_table(data=df, index='Pclass', columns='Sex', values='Survived', aggfunc='mean')
```
